```python
filtered_texts = []
```

```python
for i in range(len(texts)):
    if len(texts[i]) != 0:
        if texts[i][-1] != ")" and texts[i][0] != "(":
            filtered_texts.append(texts[i])
```

```python
article = ''

articles = []

for text in filtered_texts:
    if text.split()[0] == 'Статья' and article != '':
        articles.append(article)
        article = f"{text}. "
    else:
        article += f"{text}. "
```

```python
from simpletransformers.question_answering import QuestionAnsweringModel
from simpletransformers.question_answering import QuestionAnsweringArgs

model_args = QuestionAnsweringArgs()
model_args.n_best_size = 1
model_args.use_cuda = False
model_args.f16 = False

model = QuestionAnsweringModel(
    "roberta",
    "deepset/roberta-base-squad2",
    use_cuda = False,
    args=model_args)
```

```python
context_text = "Даня любит играть в доту."

predictions, raw_outputs = model.predict(
    [
        {
            "context": context_text,
            "qas": [
                {
                    "question": "Что любит делать Даня?",
                    "id": "0",
                }
            ],
        }
    ]
)
```
